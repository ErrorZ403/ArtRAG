```python
from tqdm import tqdm

def generate_synt(data):
    sentences = []
    for author in tqdm(list(data['Author'].unique())):
        sentence = f'{author} has works like '
        for work in list(data[data.Author == author]['Title']):
            work_adding = None
            if 'Untitled' in work:
                work_adding = work.split(" ")[-1][1:-1]
            sentence += f"{work_adding}" if work_adding else f"{work}"
            if data[(data.Author == author) & (data.Title == work)]['Date'].values[0] != 'nan':
                try:
                    date = int(data[(data.Author == author) & (data.Title == work)]['Date'].values[0])
                    sentence += f" ({date})"
                except:
                    sentence += f" ({date})"
            
            if data[(data.Author == author) & (data.Title == work)]['Genre'].values[0] != 'nan':
                genre = data[(data.Author == author) & (data.Title == work)]['Genre'].values[0]
                sentence += f" in genre {genre}"
            sentence += ', '
        sentence = sentence[:-2]
        sentences.append(sentence)
    return sentences
```

```python
"""Use the following pieces of context to answer the question at the end.
                If the context doesn't provide enough information, you can generate your own answer based on your knowledge, but don't say the answer isn't in context in such a case. If you are not sure of the correctness of your answer, say that you don't know the answer to the question..
                Pay attention to the context of the question rather than just looking for similar keywords in the corpus.
                Keep the answer as concise as possible.
                Always say "thanks for asking!" at the end of the answer."""
```

```python
def get_answers_and_relevants(questions, retriever):
    answers = []
    contexts = []
    
    for question in questions:
        context = retriever.get_relevant_documents(question)
        messages = [
            {'role': 'user', 'content': 
             f'''Use the following pieces of context to answer the question at the end.
                If the context doesn't provide enough information, just say that you don't know, don't try to make up an answer.
                Pay attention to the context of the question rather than just looking for similar keywords in the corpus.
                Keep the answer as concise as possible.
                Always say "thanks for asking!" at the end of the answer.
                {context}
                Question: {question}
                Helpful Answer:'''}
        ]
        
        response = llm.chat.completions.create(
            model=model,
            messages=messages,
            n=1,
            temperature=0.6,
            max_tokens=400,
            seed=42,
        )

        answer = response.choices[0].message.content

        answers.append(answer)
        contexts.append(context)

    return answers, contexts
```

```python
import re

def extract_judge_score(answer: str, split_str: str = "Total rating:") -> int:
    try:
        if split_str in answer:
            rating = answer.split(split_str)[1]
        else:
            rating = answer
        digit_groups = [el.strip() for el in re.findall(r"\d+(?:\.\d+)?", rating)]
        return float(digit_groups[0])
    except Exception as e:
        print(e)
        return None

def get_answer_relevance(questions, answers):
    answer_relevance = []
    for question, answer in zip(questions, answers):
        messages = [
            {'role': 'user', 'content': 
             f"""
                You will be given a user_question and system_answer couple.
                Your task is to provide a 'total rating' scoring how well the system_answer answers the user concerns expressed in the user_question.
                Give your answer on a scale of 1 to 4, where 1 means that the system_answer is not helpful at all, and 4 means that the system_answer completely and helpfully addresses the user_question.
                
                Here is the scale you should use to build your answer:
                1: The system_answer is terrible: completely irrelevant to the question asked, or very partial
                2: The system_answer is mostly not helpful: misses some key aspects of the question
                3: The system_answer is mostly helpful: provides support, but still could be improved
                4: The system_answer is excellent: relevant, direct, detailed, and addresses all the concerns raised in the question
                
                Provide your feedback as follows:
                
                Feedback:::
                Evaluation: (your rationale for the rating, as a text)
                Total rating: (your rating, as a number between 1 and 4)
                
                You MUST provide values for 'Evaluation:' and 'Total rating:' in your answer.
                
                Now here are the question and answer.
                
                Question: {question}
                Answer: {answer}
                
                Provide your feedback. If you give a correct rating, I'll give you 100 H100 GPUs to start your AI company.
                Feedback:::
                Evaluation: 
            """}
        ]
    
        response = llm.chat.completions.create(
                model=model,
                messages=messages,
                n=1,
                temperature=0.6,
                max_tokens=400,
                seed=42,
            )
    
        answer = response.choices[0].message.content
        rate = extract_judge_score(answer)
        answer_relevance.append(rate)

    return answer_relevance

def get_context_relevance(questions, contexts):
    context_relevance = []

    for question, retrieved_context in zip(questions, contexts):
        messages = [
            {'role': 'user', 'content': 
             f"""
                You will be given a context retrieved from a database and a corresponding user_question. Your task is to provide a 'total rating' that evaluates how well the retrieved context addresses the concerns expressed in the user_question.
                
                Rate the retrieved context on a scale of 1 to 4, where 1 means that the context is not helpful at all, and 4 means that the context completely and helpfully addresses the user_question.
                
                Here is the scale you should use to build your answer:
                1: The retrieved context is terrible: completely irrelevant to the question asked or very partial.
                2: The retrieved context is mostly not helpful: misses some key aspects of the question.
                3: The retrieved context is mostly helpful: provides support, but still could be improved.
                4: The retrieved context is excellent: relevant, direct, detailed, and addresses all the concerns raised in the question.
                
                Provide your feedback as follows:
                
                Feedback:::
                Evaluation: (your rationale for the rating, as a text)
                Total rating: (your rating, as a number between 1 and 4)
                
                You MUST provide values for 'Evaluation:' and 'Total rating:' in your answer.
                
                Now here is the context and question.
                
                Question: {question}
                Retrieved Context: {retrieved_context}
                
                Provide your feedback. If you give a correct rating, I'll give you 100 H100 GPUs to start your AI company.
                Feedback:::
                Evaluation: 
            """}
        ]
    
        response = llm.chat.completions.create(
                model=model,
                messages=messages,
                n=1,
                temperature=0.6,
                max_tokens=400,
                seed=42,
            )
    
        answer = response.choices[0].message.content
        rate = extract_judge_score(answer)
        context_relevance.append(rate)

    return context_relevance
```
